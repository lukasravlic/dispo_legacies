```python
import pandas as pd
import numpy as np
```

```python
tr_bmw = pd.read_excel("C:/Users/lravlic/Downloads/Transito DP_BMW_S25.xlsx",dtype={'Nº.pedido.compra':'str','CANT':'int64','Cód.N..de.parte(empa':'str','FACTURA':'str'})
```

```python
tr_bmw.rename(columns={'Tipo.pedido':'Vía','Cód.N..de.parte(empa':'Material','CANT':'Cantidad','Nº.pedido.compra':'Doc. Compra'}, inplace=True)
```

```python
cond = [tr_bmw['Vía'] == 'Courier', tr_bmw['Vía'] == 'Aéreo', tr_bmw['Vía'] == 'Marítimo']
opc = [10, 20, 90]
tr_bmw['Dias'] = np.select(cond, opc, default=0)

# Asegúrate de que 'Fecha.creación' sea tipo datetime
tr_bmw['Fecha.creación'] = pd.to_datetime(tr_bmw['Fecha.creación'])

# Calculamos fecha estimada sumando los días
fecha_estim = tr_bmw['Fecha.creación'] + pd.to_timedelta(tr_bmw['Dias'], unit='D')

# Llenamos los valores faltantes en 'ESTIMADA DISPO' con esa fecha calculada
tr_bmw['ESTIMADA DISPO'] = pd.to_datetime(tr_bmw['ESTIMADA DISPO'])
tr_bmw['FechaFinal'] = tr_bmw['ESTIMADA DISPO'].fillna(fecha_estim)
```
